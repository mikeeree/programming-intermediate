```python
import requests
import matplotlib.pyplot as plt


def track_iss():
    url = 'http://api.open-notify.org/iss-now.json'
    data_points = []

    for _ in range(100):
        response = requests.get(url)
        data = response.json()
        timestamp = data['timestamp']
        latitude = float(data['iss_position']['latitude'])
        longitude = float(data['iss_position']['longitude'])
        data_points.append((latitude, longitude, timestamp))

    return data_points

# Track ISS and get data points
data_points = track_iss()

# Extract latitude, longitude, and timestamp values for plotting
latitudes = [point[0] for point in data_points]
longitudes = [point[1] for point in data_points]
timestamps = [point[2] for point in data_points]

# Plot the ISS path using scatter plot
plt.scatter(longitudes, latitudes, c=timestamps, cmap='cool', alpha=0.7)
plt.colorbar(label='Timestamp')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('ISS Path Tracking')
plt.grid(True)
plt.show()

track_iss()
```
